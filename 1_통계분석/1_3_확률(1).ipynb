E(X) &= \sum_{i=1}^{6} x_i \cdot P(X = x_i) \\
     &= \sum_{i=1}^{6} x_i \cdot \frac{1}{6} \\

     &= \frac{1}{6} \cdot 21 \\

\begin{align*}
