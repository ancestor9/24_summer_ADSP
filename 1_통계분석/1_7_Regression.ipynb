```python
# prompt: stepfoward ë°©ì‹

import numpy as np
def forward_selection(data, target, predictors):
    remaining_predictors = [p for p in data.columns if p not in predictors]
    results = []
    for p in remaining_predictors:
        X_temp = data[predictors + [p]]
        X_temp = sm.add_constant(X_temp)
        reg = sm.OLS(target, X_temp).fit()
        results.append({'predictor': p, 'adj_r2': reg.rsquared_adj})

    best_predictor = results[np.argmax([r['adj_r2'] for r in results])]['predictor']
    predictors.append(best_predictor)

    if np.max([r['adj_r2'] for r in results]) > 0.99:
        return predictors
    else:
        return forward_selection(data, target, predictors)

predictors = ['median_income']
predictors = forward_selection(df, df['median_house_value'], predictors)
print(predictors)
```
