# **문제 75번 풀이**

## **주어진 정보**

- **전체 인구 중 해당 질병을 가지고 있는 사람의 비율 : P(질병)=0.10**
- **전체 인구 중 진단 결과 질병이 있다고 진단된 사람의 비율 : P(양성)=0.20**
- **해당 질병을 가지고 있는 사람 중 90%가 질병이 있다고 진단됨 P(양성/질병)=0.9**
- **양성진단일 경우 질병을 가질 확률 P(질병/양성) = P(양성진단과 질병의 교집합) / P(양성) = P(양성/질병) * P(질병) / P(양성) = (0,1*0.9)/0.2 = 0.09/0.2 = 0.45**

## **혼동 행렬**
### **전체 인구를 1000명이라고 가정하고**

| 실제\예측(진단)       | 양성 (Positive) | 음성 (Negative) | Support |
|-----------------|------------------|------------------|------------------|
| 질병 (Positive) | TP  (90)         | FN (10)      | (100)               |
| 정상 (Negative) | FP  (110)        | TN  (790)    | (900)               |
|                 | 200              | 800            | 1000            |

- **TP (True Positive)**: 실제 양성이고 양성으로 예측된 경우
- **FN (False Negative)**: 실제 양성인데 음성으로 예측된 경우
- **FP (False Positive)**: 실제 음성인데 양성으로 예측된 경우
- **TN (True Negative)**: 실제 음성이고 음성으로 예측된 경우

### **관심있는 양성을 분별하는 경우에 2가지 평가지표**
- **양성진단인 경우 질병을 가질 비율은 정밀도(Precision) = 90/200 =0.45**
- **모든 질병자인 경우 양성진단 비율 재현율/검측율(Recall) = 90/100 =0.90**
