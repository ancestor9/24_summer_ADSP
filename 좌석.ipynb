```python
#     2    ()
import cv2
import matplotlib.pyplot as plt

#   
image_path = 'cx-insight-YloghyfD7e8-unsplash.jpg'

#  
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Haar Cascade Classifier   
#     : https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

#  
faces = face_cascade.detectMultiScale(image_rgb, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

#    2  
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.title('Detected Faces')

#   
for (x, y, w, h) in faces:
    plt.gca().add_patch(plt.Rectangle((x, y), w, h, fill=False, color='red', linewidth=2))

plt.show()

# 2    
plt.figure(figsize=(10, 6))
plt.scatter([x + w//2 for (x, y, w, h) in faces], [y + h//2 for (x, y, w, h) in faces], color='red')
plt.xlim(0, image_rgb.shape[1])
plt.ylim(image_rgb.shape[0], 0)
plt.title('Face Positions in 2D Plane')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(True)
plt.show()
```
